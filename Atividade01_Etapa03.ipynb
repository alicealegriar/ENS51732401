```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Criando o dataframe
df = pd.DataFrame({
    "Mês": ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"],
    "Estimativa da Produção (toneladas)": [233333, 233333, 233333, 233333, 233333, 233333, 233333, 233333, 233333, 233333, 233333, 233333]
})

# Definindo o índice do dataframe
df.set_index("Mês", inplace=True)

# Criando o gráfico
plt.plot(df.index, df["Estimativa da Produção (toneladas)"])
plt.xlabel("Mês")
plt.ylabel("Estimativa da Produção (toneladas)")
plt.title("Estimativa da Produção Mensal de Cimento da Cimentos Itambé em 2022")
plt.show()
```
