```python
# criando uma lista em python
list = [0, 'eu amo polAtm', 2.5]
print(list)

# Nesta parte fiz a adicao entre duas strings, sendo que list[1] eh o segundo indice do vetor list
print('variavel:'+ list[1])

# Adicao do primeiro indice da variavel list (list[0]) - Lembrar que a indexacao no python inicia em 0
print(list[0] + 1)
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print(a)
print('indexacao numpy')
print(a[0,0])
```

```python
# Criando numeros aleatorios e mostrando a dimensao da matriz
b = np.random.rand(5,5)
print(b)
print('Dimensao da matriz')
print(b.shape)
```

```python
# Criando matriz de valores aleatorios entre uma faixa estabelecida
vmin = 1 # valor minimo
vmax = 100 # valor maximo

# Matriz 
c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c)
```

```python
# PLotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
plt.title('Graficos com Valores Inteiros no Eixo X')
plt.xticks(x) # Definindo os valores do eixo x
```

```python
# Criando um grafico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# Exportar a figura
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\,'/') + '/figures/'
print(figFolder)
fig.savefig('/Figure01.png')
```
